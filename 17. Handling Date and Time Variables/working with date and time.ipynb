# Working with Dates

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

### 1. Extract year

```python
date['date_year'] = date['date'].dt.year

date.head()
```

### 2. Extract month

```python
date['date_month'] = date['date'].dt.month

date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

### 3. Extract days

```python
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week
date['date_dayofweek'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week - name

date['date_dayofweek_name'] = date['date'].dt.day_name()

date.drop(columns = ['product_id','city_id','orders']).head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dayofweek_name'].isin(['Sunday','Saturday']),1,0)

date.drop(columns = ['product_id','city_id','orders']).head()
```

### 4. Extract quarter

```python
date['quarter'] = date['date'].dt.quarter
date.drop(columns = ['product_id','city_id','orders']).head()
```

### 5. Extract semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns = ['product_id','city_id','orders']).head()
```

### 6. Extract time elapsed between two dates

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```
